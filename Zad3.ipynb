```python
import pandas as pd
stragan = pd.read_csv('Straganik.csv')
```

```python
def VegeFinder(item):
    data = []
    for row in range(len(stragan)):
        rowSum = 0
        for x in range(len(stragan.loc[row])-1):
            rowSum+=stragan.iloc[row,x]*item[x]
        data.append(round(rowSum,3))
    maxPoints = max(data)
    return stragan.at[data.index(maxPoints),"nazwa"]

Szukam = [0.9, 0.0, 0.0, 0.8, 0.8, 0.0, 0.0, 0.7, 0.0, 0.0]
VegeFinder(Szukam)
```
