```python
import pandas as pd
import seaborn as sns
import random
import numpy as np
import math
penguins = pd.read_csv('penguins.csv')
```

```python
class ProcessingData:
    
    @staticmethod
    def shuffle(base):
        for x in range(len(base)-1,0,-1):
            newPlace = random.randint(0,x-1)
            base.loc[x],base.loc[newPlace]=base.loc[newPlace],base.loc[x]
            
    @staticmethod
    def normalize(base):
        base.iloc[:,2:-1] = (base.iloc[:,2:-1]-base.iloc[:,2:-1].min())/(base.iloc[:,2:-1].max()-base.iloc[:,2:-1].min())

    @staticmethod
    def devide(base):
        n = int(len(base)*0.7)
        base_tren = base[:n]
        base_valid = base[n:]
        return(base_tren,base_valid)
```

```python
penguinsEdit = penguins.copy()

ProcessingData.shuffle(penguinsEdit)
ProcessingData.normalize(penguinsEdit)
trening_data,validate_data = ProcessingData.devide(penguinsEdit)
```

```python
class KNN:

    def distance(x, y, m):
        dist = 0
        for i in range(2,len(y)-1):
            dist += abs(x[i]-y[i])**m
        dist = float(dist**(1/m))
        return dist

    @staticmethod   
    def GetDistance(v):
          return v[1]

    @staticmethod
    def clustering(validate_data, training_data, k, m):
        correct = 0
        for sampleId in range(len(validate_data)):
            distances = list()
            sample = validate_data.iloc[sampleId]
            trueName = sample.species
            for i in range(len(training_data)):
                distance = KNN.distance(sample,training_data.iloc[i],m)
                distances.append((training_data.iloc[i],distance))
                
            distances.sort(key=KNN.GetDistance)
            
            classes = {'Adelie':0,'Chinstrap':0,'Gentoo':0}
            for i in range(k):
                classes[distances[i][0].species] +=1

            if trueName ==  max(classes, key=classes.get):
                correct+=1
        return correct/len(validate_data)*100
```

```python
m = 3
for k in range (2,5):
    print("k={} m={} => Dokładność - {:.2f}%".format(k,m,KNN.clustering(validate_data,trening_data,k,m)))
```
